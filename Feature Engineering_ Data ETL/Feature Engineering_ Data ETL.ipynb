````python
# Determine the number of null values in each column
nan_sum = df_bfrss2021.isnull().sum()
pd.options.display.max_rows = 4000

print(nan_sum)
````

````python
# Percentage of missing values
nan_mean = df_bfrss2021.isnull().mean()
pd.options.display.max_rows = 4000

print(nan_mean)
````

The majority of rows and columns include missing data.  As the dataset reflects the results if a comprehensive survey, one would expect to find a large number of missing data, and that the nature of the missing data would be highly variable and random.  Deleting rows or columns would remove most indicators.  The best approach is to fill in missing data.  Missing values are of the MCAR type, meaning there is absolutely no relationship between the data missing and any other observed or missing value in the dataset. In other words, those missing data points are a random subset of the dataset.  The most effective approach is to use a random sample imputer (using median) to fill in the missing data.


````python
# Fill in NaN values with Random Sampler Imputation by Median
df_bfrss2021.fillna(df_bfrss2021.median())
````
