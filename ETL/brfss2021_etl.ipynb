## Feature Selection
Finding the correlation between the variables in our data and the outcome variable with Ridge Model

````python
# Add the values to an array
array = brfss_final.values

X = array[:,0:27]
Y = array[:,0]
````

````python
# First things first
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(X,Y)
````

````python
Ridge(alpha=1.0, copy_X=True, fit_intercept=True, max_iter=None,
   random_state=None, solver='auto', tol=0.001)
````

````python
# A helper method for pretty-printing the coefficients
def pretty_print_coefs(coefs, names = None, sort = False):
    if names == None:
        names = ["X%s" % x for x in range(len(coefs))]
    lst = zip(coefs, names)
    if sort:
        lst = sorted(lst,  key = lambda x:-np.abs(x[0]))
    return " + ".join("%s * %s" % (round(coef, 3), name)
                                   for coef, name in lst)
````

````python
print ("Ridge model:", pretty_print_coefs(ridge.coef_))
````

The following points should be kept in mind when applying Ridge regression:

It is also known as L2-Regularization.

For correlated features, it means they tend to produce similar coefficients.

Features with negative coefficients do not contribute as much as those with positive coefficients. However, in a more complex scenario where many features are involved, this score will certainly assist in the final selection of features.

From the Ridge Model Above, the features with positive correlation contribute more to the output variable and will help in feature selection decision making process:

X6 - Heart Attack, X7 - Heart Disease, X8 - Stroke, X10 - Kidney Disease, X11 - Difficulty Walking, X12 - Race/Ethnicity, X13 - Overall Health, X14 - Health Plan, X16 - High BP, X18 - Age, X19 - Overweight/Obese, X22 - Smoker, X25 - Fruit>1/Day, X26 - Veggies>1/Day
